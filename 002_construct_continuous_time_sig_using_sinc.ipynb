In this notebook we look at reconstructing a continouts-time signal from its corresponding discrete-time signal.
We simulate a continuous-time signal, with a discrete-time signal with a low sampling time (high sampling rate).

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
Fsc = 100; # sampling frequency for simulating the continuous-time
Tsc = 1.0 / Fsc # sampling time-step for simulating the continuous-time
tf = 1.11;
tc = np.arange(0, tf, Tsc) # continuous-time
```

```python
f = 1 # Hz # frequency of the continuous-time signal
xc = np.sin(2*np.pi*f*tc) # continuous-time signal
```

```python
Ts = 2*Tsc
Fs = 1./Ts # sampling frequency for the descrete-time signal
td = np.arange(0, tf, Ts)
```

```python
xd = np.sin(2*np.pi*f*td)
```

```python
plt.plot(tc, xc)
plt.plot(td, xd, 'o')
plt.legend(['continuous-time signal', 'discrete-time signal'])
```

```python
xc_constructed = 0.0 * tc
for n in range(len(xd)):
    arg = (tc-n*Ts)/Ts
    xc_constructed = xc_constructed + xd[n] * np.sinc(arg)#/arg#
    
```

```python
plt.plot(tc, xc, 'b')
plt.plot(tc, xc_constructed, 'r')
plt.legend(['continuous-time signal', 'constructed (from discrete samples) continuous-time signal'])
```

```python
plt.plot(tc, xc-xc_constructed, '*')
plt.legend(['error between continous-time signal and constructed signal'])
```
